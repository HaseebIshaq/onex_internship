```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
newdf = pd.read_csv("sensor.csv")
newdf
```

```python
# duplicate removal
newdf=newdf.drop_duplicates()
newdf=newdf.dropna()
```

```python
newdf['Temperature'].describe()
```

```python
newdf['Humidity'].describe()
```

```python
fig = px.line(newdf, x='Timestamp', y='Humidity', title='Temperature and Humidity over time', color_discrete_sequence=['red'])
fig.show()
```

```python
fig2 = px.line(newdf, x = 'Timestamp', y = 'Temperature', title='Temperature and Humidity over time',color_discrete_sequence=['blue'])
fig2.show()
```

```python

newdf['Timestamp']=pd.to_datetime(newdf['Timestamp'])
newdf['Timestamp']

def get_time_Category(hour):
    if 6<=hour<12:
        return 'Morning'
    elif 12<=hour<18:
        return 'Afternoon'
    elif 18<=hour <24:
        return 'Evening'
    else:
        return 'Night'

newdf['Category']=newdf['Timestamp'].dt.hour.apply(get_time_Category)
  
fig3= px.scatter(newdf, x='Temperature', y='Humidity',color='Category')
fig3.show()
```

```python
newdf.set_index('Timestamp', inplace=True)
fig = px.imshow(newdf, text_auto=True)
fig.show()

```
